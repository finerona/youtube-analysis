## creating new column Average_view

```python

newdf['Average_view'] =round( newdf['total_views'] / newdf['total_videos'])
```

```python
newdf.head()
```

## sorting the values from ascending order for total_views

```python
newdf.sort_values(by=['total_views'],ascending=False).head()
```

## sorting the values from ascending order for total_videos

```python
newdf.sort_values(by=['total_videos'],ascending=False).head()
```

## sorting the values from ascending order for total_subscribers

```python
newdf.sort_values(by=['total_subscribers'],ascending=False).head()
```

```python
convertion_rate = 0.02
cost_per_item = 2 #dollars
campaign_cost = 10000 #dollars 
```

```python
newdf['units_sold_per_video'] = newdf['Average_view'] * convertion_rate
```

```python
newdf.head()
```

```python
newdf.sort_values(by=['units_sold_per_video'],ascending=False).head()
```

```python
## finding the revenue_per_video 
```

```python
newdf['revenue_per_video'] = newdf['Average_view'] * convertion_rate * cost_per_item
```

```python
newdf.head()
```

```python
newdf.sort_values(by=['revenue_per_video'],ascending=False).head()
```

```python
## finding the net profit
```

```python
newdf['net_profit'] = newdf['Average_view'] * convertion_rate * cost_per_item - campaign_cost
```

```python
newdf.head()
```

```python
newdf.head()
```

```python
newdf.sort_values(by=['net_profit'],ascending=False).head()
```

```python
## finding the net_profit_by_subscribers
```

```python
newdf['net_profit_by_subscribers'] = newdf['total_subscribers'] * convertion_rate * cost_per_item - campaign_cost
```

```python
newdf.head()
```

```python
## sorting the net_profit_by_subscribers by ascending
```

```python
newdf.sort_values(by=['net_profit_by_subscribers'],ascending=False).head()
```

```python
## finding the netprofit_by_videos
```

```python
newdf['netprofit_by_total_vidoes'] = newdf['total_videos'] * convertion_rate * cost_per_item - campaign_cost
```

```python
## sort the netprofit_by_total_vidoes by ascending
```

```python
newdf.sort_values(by=['netprofit_by_total_vidoes'],ascending=False).head()
```

```python
##saving the project to csv file on local device
```
